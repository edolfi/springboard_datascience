# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

I believe a more meaningful analysis should include a broader range of hospital sizes. The thresholds for hospital size used in the analysis above only consider 10% of all hospitals in the dataset, which won't mean very much if trying to influence policy decisions. Instead, I will use the following thresholds - less than 150 discharges = small hospitals and more than 500 discharges = large hospitals. Therefore, hospitals belows the 25th percentile of # of discharges are small, hospital sizes within the IQR are medium-sized, and hospitals > the 75th percentile are large. (I obviously rounded the percentiles!)

<b>Null hypothesis:</b> Small and large hospitals do not differ in their rates of readmissions.

<b> Alternative hypothesis:</b> Smaller hospitals have higher rates of readmissions.

```python
clean_hospital_read_df.dropna(subset=['Number of Discharges','Excess Readmission Ratio'], inplace=True)
small_hosp = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] <= 150]['Excess Readmission Ratio']
large_hosp = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] >= 500]['Excess Readmission Ratio']
print(np.mean(small_hosp), np.mean(large_hosp))
```

```python
# generate permutation samples to compare difference in means
def diff_of_means(data_1, data_2):
    """Difference in means of two arrays."""

    # The difference of means of data_1, data_2: diff
    diff = np.mean(data_1) - np.mean(data_2)

    return diff

# Compute difference of means from sample: empirical_diff_means
sample_diff_means = diff_of_means(small_hosp, large_hosp)
print(sample_diff_means)
```

```python
def permutation_sample(data1, data2):
    """Generate a permutation sample from two data sets."""

    # Concatenate the data sets: data
    data = np.concatenate((data1,data2))

    # Permute the concatenated array: permuted_data
    permuted_data = np.random.permutation(data)

    # Split the permuted array into two: perm_sample_1, perm_sample_2
    perm_sample_1 = permuted_data[:len(data1)]
    perm_sample_2 = permuted_data[len(data1):]

    return perm_sample_1, perm_sample_2

def draw_perm_reps(data_1, data_2, func, size=1):
    """Generate multiple permutation replicates."""

    # Initialize array of replicates: perm_replicates
    perm_replicates = np.empty(size)

    for i in range(size):
        # Generate permutation sample
        perm_sample_1, perm_sample_2 = permutation_sample(data_1,data_2)

        # Compute the test statistic
        perm_replicates[i] = func(perm_sample_1,perm_sample_2)

    return perm_replicates

# Acquire permutation samples
perm_replicates = draw_perm_reps(small_hosp, large_hosp, diff_of_means, 1000)
print(np.percentile(perm_replicates, [0.5,99.5]))

# Compute and print p-value: p
p = np.sum(perm_replicates >= sample_diff_means) / len(perm_replicates)
print('p-value = '+str(p))
```

**Report statistical significance for  α  = .01.**
- From the above p-value, we can reject the null hypothesis that hospital size, i.e. number of discharges, does not affect the rate of readmissions in hospitals.
